```python
%sql
CREATE TABLE IF NOT EXISTS lakehousecat1.deltadb.sales_orders_liquid
(
  order_id INT,
  customer_id INT,
  region STRING,
  product STRING,
  quantity INT,
  price DOUBLE,
  order_date DATE
)
USING DELTA
CLUSTER BY (customer_id, region);
```
