```python
%sql
CREATE OR REPLACE TABLE lakehousecat1.deltadb.tblsales
(
  sales_id INT,
  product_id INT,
  region STRING,
  sales_amount DOUBLE,
  sales_date DATE
)
USING DELTA;
```
